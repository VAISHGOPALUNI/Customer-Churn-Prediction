```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
dataset["State"]= le.fit_transform(dataset["State"])
```

```python
x = dataset.iloc[:,0:4].values
y = dataset.iloc[:,4].values
```

```python
x
```

```python
from sklearn.preprocessing import OneHotEncoder
one = OneHotEncoder()
z = one.fit_transform(x[:,3:4]).toarray()
x = np.delete(x,3,axis = 1)
x = np.concatenate((z,x),axis = 1)
```

```python
x.shape
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2,random_state = 0)
```

```python
#ANN Starts
from keras.models import Sequential
from keras.layers import Dense
```

```python
model = Sequential()
```

```python
model.add(Dense(units = 6 , init = "random_uniform",activation = "relu"))
```

```python
model.add(Dense(units = 12 , init = "random_uniform",activation = "relu"))
```

```python
model.add(Dense(units = 24 , init = "random_uniform",activation = "relu"))
```

```python
model.add(Dense(units = 21 , init = "random_uniform",activation = "relu"))
```

```python
model.add(Dense(units = 9 , init = "random_uniform",activation = "relu"))
```

```python
model.add(Dense(units = 1 , init = "random_uniform"))
```

```python
model.compile(optimizer = "adam",loss = "mean_squared_error",metrics = ["mean_squared_error"])
```

```python
model.fit(x_train,y_train,epochs = 2000,batch_size = 8)
```

```python
y_pred = model.predict(x_test)
```
